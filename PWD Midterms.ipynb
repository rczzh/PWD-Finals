## 2.1 Defining scraping and extraction functions

The list of URLs to be analysed during this project are obtained through a CSV file. Before we import and retrieve the contents of the URLs, we must first define a function that we can call to extract and store the data as we iterate through the URLs.

The below function checks if the webpage is accessible and returns the contents if possible:

```python
def getParsedWebpage(url):
    
    # Verifies the page is accessible for scraping
    
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:12.0) Gecko/20100101 Firefox/12.0'}
    
    req = requests.get(url, headers = headers)
    
    if req.status_code != 200:
        soup = 'error'
    else:
        content = req.content
        
        # Parses the webpage if its accessible
        soup = BeautifulSoup(req.content, 'html.parser')
        
    return soup
```

### Extracting information from IGN webpages

The following function extracts review verdict text from a IGN webpage:

```python
def extractInfoIGN(soup):
    
    # Extracts game title and removes extra words in the headlines
    title = soup.find("h1").get_text().replace(" Review", "")
    
    # Extracts review text
    verdict = soup.find_all("div", {"class": "verdict"})
    
    # Removes extra html elements and 
    # compiles the verdict text into a single string that is seperated by 1 space between each word
     
    verdictText = ""
    for content in verdict:
        verdictText = verdictText + ' ' + content.get_text().lstrip().rstrip()
        
    return title, verdictText[8:]
```

### Extracting information from steamchart webpages

The following function extracts the following from a steamchart webpage:
- Game release date
- Peak player count to date
- Player count on release date
- Player count 6 months after release date

```python
def extractInfoSteamChart(soup):
    
    # Index set to -1 to retrieve last element in list
    i = -1
    
    # Extracts release date
    releaseDate = soup.find_all("td", {"class" : "month-cell"})[i].get_text()  

    # Extracts peak player count
    peakPlayerCount = soup.find_all("div", {"class" : "app-stat"})[-1].get_text()
    
    # Removes extra words, leaving behind numbers only
    peakPlayerCount = re.sub('\D', '', peakPlayerCount)
    
    # Extracts player count on release date
    lastTableEntryPeakPlayer = soup.find_all("td", {"class" : "right num"})[i].get_text()
    
    # Removes trials player count entries
    while int(lastTableEntryPeakPlayer) < 100:
        i -= 1
        lastTableEntryPeakPlayer = soup.find_all("td", {"class" : "right num"})[i].get_text()

    # Extracts player count 6 months after release date
    futureDatePlayerCount = soup.find_all("td", {"class" : "right num"})[i - 6].get_text()  
    
    return releaseDate.strip(), peakPlayerCount, lastTableEntryPeakPlayer, futureDatePlayerCount
```

## 2.2 Import URLs and scrape

We will convert the CSV of URLs into a dataframe then into a list and attempt to scrape the information of each of the URLs we have imported. If successful, we will append to a newly created dataframe for scraped information.

```python
ignUrls = pd.read_csv("ignURLs.csv")
ignUrlDf = pd.DataFrame(ignUrls)
ignUrlList = ignUrlDf.values.tolist()

steamChartUrls = pd.read_csv("steamChartURLs.csv")
steamChartUrlDf = pd.DataFrame(steamChartUrls)
steamChartList = steamChartUrlDf.values.tolist()

scrapedIgnData = pd.DataFrame(columns=["Game Title", "Verdict Text"])
scrapedSteamChartData = pd.DataFrame(columns=["Release Date", "Player Count on Release",
                                              "Player Count After 6 Months", "Peak Player Count"])

for i in range(0, len(ignUrlList)):
    
    # Check webpage is accessible and get parsed webpage, if not accessible soup will be returned as 'error'
    ignSoup = getParsedWebpage(ignUrlList[i][0])
    steamSoup = getParsedWebpage(steamChartList[i][0])

    if ignSoup == 'error':
        print("Error fetching webage, data not scraped: " + ignUrlList[i][0])
    elif steamSoup == 'error':
        print("Error fetching webage, data not scraped: " + steamChartList[i][0])
        continue
        
    title, verdictText = extractInfoIGN(ignSoup)
    releaseDate, peakPlayerCount, lastTableEntryPeakPlayer, futureDatePlayerCount = extractInfoSteamChart(steamSoup)
    
    reviewInfo = pd.Series({'Game Title': title, 'Verdict Text': verdictText})
    scrapedIgnData = pd.concat([scrapedIgnData, reviewInfo.to_frame().T], ignore_index=True)
               
    steamChartInfo = pd.Series({'Release Date': releaseDate, 'Peak Player Count': peakPlayerCount,
                                'Player Count on Release': lastTableEntryPeakPlayer, "Player Count After 6 Months": futureDatePlayerCount})
    scrapedSteamChartData = pd.concat([scrapedSteamChartData, steamChartInfo.to_frame().T], ignore_index=True)

# Merge both dataframes
scrapedCombinedData = pd.concat([scrapedIgnData, scrapedSteamChartData], axis = 1)
```

## 3.2 Removing stop words

We need a version of a corpus without stopwords. As we will be analysing the frequency of words to obtain a description of the game, we ought to remove stop words as they would appear with frequently despite not being meaningful to our analysis.

### 3.2.1 Identifying stop words not covered by NLTK's stop words

We will be using NLTK to remove stop words. However, NLTK might not be comprehensive enough to remove all possible stop words. As such we will be analysing the top few frequent words in each of our year corpus to see what are some of the common words.

```python
# We will firstly define a function to retrieve the word frequency of each corpus
# We will also remove stop words using NLTK's stopword set

nltk.download('stopwords')

def getFreq(corpus):
    
    # Removing certain punctuations using regrex. Do not remove apostrophes
    corpus = re.sub('\.|\,|\"|\“|\”|\—|\–|\-|\?|\!|\:|\;|\(|\)', '', corpus)
    
    # Split the corpus into a list of words
    corpusList = corpus.lower().split()
  
    # Converting to a set means all duplicate words are removed
    uniqueWordsSet = set(corpusList)
    
    # Removing stop words
    uniqueWords = [word for word in uniqueWordsSet if not word in stopwords.words('english')]
    
    # Creates new dataframe to store our word-frequency pairs
    wordFreq = pd.DataFrame(columns = ['Word', 'Frequency'])
    
    for word in uniqueWords:
        wordFreqPair = pd.Series({"Word" : word, 'Frequency': corpusList.count(word)})
        wordFreq = pd.concat([wordFreq, wordFreqPair.to_frame().T], ignore_index=True)
        
    return wordFreq
    
```

We will now use the above function to get the word frequencies of each yearly corpus and access if we would like to treat any of the most frequent words as stop words.

### 3.2.2 Create filtered corpora

We will be defining another function to remove our new set of stopwords from each of the year corpus:

```python
nltk.download('punkt')

def removeStopWords(corpus):
    corpus = corpus.lower()
    
    # Removing certain punctuations using regrex. Do not remove apostrophes
    corpus = re.sub('\.|\,|\"|\“|\”|\—|\–|\-|\?|\!|\:|\;|\(|\)', '', corpus)
    
    # Tokenize the corpus - this creates a list of all the words in the corpus 
    tokens = word_tokenize(corpus)
    
    # Apart from stopwords, we will remove tokens with only punctuation or suffixes
    punct = ["'", "''", '``', '(', ')', '%', '&', '...', '…', "‘", "’"]
    suffixes = ["'s", "n't", "'ve", "'ll", "'re", "'d"]
    removeWords = punct + suffixes + customStopWords + stopwords.words('english')
    
    # Create a new list with stop words removed from the corpus
    filteredTokens = [word for word in tokens if not word in removeWords]
    
    filteredCorpus = ' '.join(filteredTokens)
    
    return filteredCorpus
```

## 3.3 Lemmatization

We will 'distill' the words we have into their root form to make them more meaningful for analysis. In our analysis later on, we would want to retrieve descriptive terms for the each corpus.

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()

def lemmatizeCorpus(corpus):
    tokens = word_tokenize(corpus)

    lemmatizedTokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    return ' '.join(lemmatizedTokens)
```

```python
# Lemmatize the Filtered Aggregate Verdict Text and add as a new column
verdictClusterDf['Lemmatized Aggregate Verdict Text'] = ''

for i in range(0, len(verdictClusterDf)):
    verdictClusterDf.loc[i, 'Lemmatized Aggregate Verdict Text'] = lemmatizeCorpus(verdictClusterDf['Filtered Aggregate Verdict Text'][i])
```

By extracting the descriptive terms of each corpus throughout the year, we are able to formulate a general consensus of the state of those games released in that particular year. 

```python
nltk.download('averaged_perceptron_tagger')

# Retrieve descriptive words via tags and add as a new column
verdictClusterDf['Descriptors'] = ''

# Function to tag each individual word with its specific tag using NLTK tagger
def tagWord(corpus):
    tagWords = ""
    
    # Taglist will only retrieve words with 'JJ' tag, which stands for adjectives
    tagList = [word for word,tag in nltk.pos_tag(nltk.word_tokenize(corpus)) if tag[0] in "JJ"]
    
    # Combine words in list into a single string
    for word in tagList:
        tagWords = tagWords + " " + word
        
    return tagWords
        
for i in range(0, len(verdictClusterDf)):
    verdictClusterDf.loc[i, 'Descriptors'] = tagWord(verdictClusterDf['Filtered Aggregate Verdict Text'][i])
```
